```python
l = list(df['name'].unique())
l.sort()

output={}
for i in range(0,len(l)):
    name = l[i]
    tempdf = df[df['name'] == name].reset_index(drop=True)
    
    for j in range(0,len(tempdf)):
        
        date = tempdf['date'][j]
        data = tempdf['data'][j]
        
        try:
            output[name].append({'x':date,'y':data})
        except:
            output[name] = [{'x':date,'y':data}]
```
